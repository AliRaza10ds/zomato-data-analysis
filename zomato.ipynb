```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set= pd.read_csv('Zomato data .csv')
data_set
```

```python
#Types of restaurant and number of orders
type_of_reataurant=data_set['listed_in(type)'].value_counts()
plt.plot(type_of_reataurant)
plt.title('Types of restaurant and number of orders')
plt.ylabel('number of orders')
plt.xlabel('types of restaurant')          
```

```python
#votes received by the restaurants from customers
plt.bar(data_set['listed_in(type)'],data_set['votes'])
plt.title('votes received by the restaurants from customers')
plt.ylabel('no of votes')
plt.xlabel('types of restaurant')
plt.show()        
        
```

```python
#top10 rated restaurants
top_rating=data_set[['name' ,'rate']]
top_ratings=top_rating.sort_values(by='rate',ascending=False).reset_index()
top_ratings.head(10)
```
